```python
import pandas as pd
import json

# 读取 CSV
df = pd.read_csv('cardDB.CSV', encoding='GBK')

# 配置
group_column = '职业' # 分组列
name_column = '中文名' # 要提取的列
priority_order = ["Ironclad"] # 分组优先级

# 自定义排序函数
def sort_key(group_name):
    if group_name in priority_order:
        return (0, priority_order.index(group_name))
    return (1, group_name)

# 分组
grouped = df.groupby(group_column)
sorted_groups = sorted(grouped.groups.keys(), key=sort_key)

# 按顺序提取名字，生成 {1: "名字1", 2: "名字2", ...}
result = {}
index = 1
for group_name in sorted_groups:
    group_data = grouped.get_group(group_name)
    for name in group_data[name_column]:
        result[index] = name
        index += 1

# 保存 JSON
with open('card.json', 'w', encoding='utf-8') as f:
    json.dump(result, f, ensure_ascii=False, indent=2)

print(f"共提取 {len(result)} 条数据")
```
